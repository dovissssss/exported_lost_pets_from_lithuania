```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sb

DIRECTORY = "../data/raw/Prarasti_isvezti_augintiniai.csv"

df = pd.read_csv(
    DIRECTORY,
    delimiter=";",
    dtype={
        " Išvežimo, praradimo pavadinimas": "string",
        " Šalies pavadinimas": "string",
        " Gyvūnų augintinio laikymo vietos savivaldybės pavadinimas": "string",
        " Gyvūnų augintinio laikymo vietos seniūnijos pavadinimas": "string",
        " Gyvūnų augintinio laikymo vietos vietovės pavadinimas": "string",
        " Gyvūno augintinio rūšies pavadinimas": "string",
        " Išvežimo praradimo metai": "int64",
        " Gyvūnų augintinių skaičius": "int64",
    },
)

display(df)
```

```python
# Exported Pets sum by years, type
df[df["Išvežimo, praradimo pavadinimas"] == "Išvežimas"].groupby(
    [" Išvežimo praradimo metai", " Gyvūno augintinio rūšies pavadinimas"],
    as_index=False,
)[" Gyvūnų augintinių skaičius"].sum()
```

```python
# Lost Pets sum by years, type
df[df["Išvežimo, praradimo pavadinimas"] == "Dingimas"].groupby(
    [" Išvežimo praradimo metai", " Gyvūno augintinio rūšies pavadinimas"],
    as_index=False,
)[" Gyvūnų augintinių skaičius"].sum()
```

```python
# Exported/Lost Pets Total Sum throughout years
df.groupby(
    [" Gyvūno augintinio rūšies pavadinimas", "Išvežimo, praradimo pavadinimas"],
    as_index=False,
)[" Gyvūnų augintinių skaičius"].sum()
```

```python
# Total exported Pets by export countries
df[df["Išvežimo, praradimo pavadinimas"] == "Išvežimas"].groupby(
    " Šalies pavadinimas", as_index=False
)[" Gyvūnų augintinių skaičius"].sum().sort_values(
    " Gyvūnų augintinių skaičius", ascending=False
)
```

```python
# Lithuanian municipalities by total exported pets sum including all analysis period
df[df["Išvežimo, praradimo pavadinimas"] == "Išvežimas"].groupby(
    " Gyvūnų augintinio laikymo vietos savivaldybės pavadinimas", as_index=False
)[" Gyvūnų augintinių skaičius"].sum().sort_values(
    " Gyvūnų augintinių skaičius", ascending=False
)
```

```python
# To how many unique countries, each year pets are exported
df[df["Išvežimo, praradimo pavadinimas"] == "Išvežimas"].groupby(
    " Išvežimo praradimo metai", as_index=False
)[" Šalies pavadinimas"].nunique()
```

```python
# Exported Pets sum, max entries and mean entries per year
df[df["Išvežimo, praradimo pavadinimas"] == "Išvežimas"].groupby(
    " Išvežimo praradimo metai", as_index=False
)[" Gyvūnų augintinių skaičius"].agg(["sum", "max", "mean"])
```

```python
# Pandas Plot()
exported_pets_by_year = (
    df[df["Išvežimo, praradimo pavadinimas"] == "Išvežimas"]
    .groupby(" Išvežimo praradimo metai", as_index=False)[" Gyvūnų augintinių skaičius"]
    .agg(["sum"])
)
exported_pets_by_year.plot(
    legend=False,
    figsize=(5, 3),
    c="green",
    marker="x",
    title="Exported Pets From Lithuania",
)
```

```python
unique_countries = (
    df[df["Išvežimo, praradimo pavadinimas"] == "Išvežimas"]
    .groupby([" Išvežimo praradimo metai"], as_index=True)[" Šalies pavadinimas"]
    .nunique()
)
unique_countries.plot.bar(legend=False, figsize=(5, 3))
```

```python
# Plot Lost Pets sum by years, type
dfr = (
    df[df["Išvežimo, praradimo pavadinimas"] == "Išvežimas"]
    .groupby(
        [" Išvežimo praradimo metai", " Gyvūno augintinio rūšies pavadinimas"],
        as_index=False,
    )[" Gyvūnų augintinių skaičius"]
    .sum()
)

sb.set(rc={"figure.figsize": (10, 5)})

sb.barplot(
    x=dfr[" Išvežimo praradimo metai"],
    y=dfr[" Gyvūnų augintinių skaičius"],
    hue=dfr[" Gyvūno augintinio rūšies pavadinimas"],
    palette="dark",
)

plt.show()
```
